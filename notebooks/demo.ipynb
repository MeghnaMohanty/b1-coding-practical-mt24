# Simulate closed-loop
After implementing your control functionality, you can simulate the closed-loop with code that looks something like this...

```python
# Import relevant modules
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from dynamic import Submarine, ClosedLoop, Mission, Trajectory  
from control import PDController  

# Instantiating the submarine and controller
sub = Submarine()
controller = PDController()  

# Instantiate the closed loop system
closed_loop = ClosedLoop(sub, controller)

# Loading the mission data
file_path = os.path.join('..', 'data', 'mission.csv')
mission = Mission.from_csv(file_path)

# Simulate the mission with random disturbances
trajectory = closed_loop.simulate_with_random_disturbances(mission)

# Plotting the completed mission
trajectory.plot_completed_mission(mission)

```
